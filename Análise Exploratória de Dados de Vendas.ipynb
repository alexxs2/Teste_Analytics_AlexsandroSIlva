# Bibliotecas necessárias

```python
#comando para instalar a biblioteca pandas  ! pip install pandas 
#comando para instalar a biblioteca matplotlib  ! pip install matplotlib  
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregando o dataset
df = pd.read_csv('data_clean.csv', sep=",", encoding='latin1')
```

# Gráfico de linhas das tendência de Vendas Mensais

```python
df['Data'] = pd.to_datetime(df['Data'], errors='coerce')


df['vendas'] = df['Quantidade'] * df['Preço']

# Agrupar os dados por mês e somar as vendas para cada Mês
df_mes = df.groupby(df['Data'].dt.to_period('M'))['vendas'].sum().reset_index()


# Plotar o gráfico de linha
plt.figure(figsize=(12,6))
plt.plot(df_mes['Data'].astype(str), df_mes['vendas'], marker='o', linestyle='-', color='b')

# título e nomeando o eixo y e o eixo x do gráfico
plt.title('Tendência de Vendas Mensais')
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.xticks(rotation=45)
plt.grid(True)

# imprimi o gráfico
plt.tight_layout()
plt.show()
```

# Identificação de Padrões do Gráfico

- Primeiro Padrão: Vemos um começo de ano moderado e crescendo com picos no mês 4 e uma queda no mês 5 porém após isso temos um crescimento acentuado atingindo seu pico no mês 7 sendo o maior deles nesse ano de 2023.

- Segundo Padrão: Após isso vemos uma queda constante até o mês 12 sendo ele o ponto mais baixo do ano inteiro, sendo o meio do ano nosso ponto mais alto e o final dele o ponto mais baixo
