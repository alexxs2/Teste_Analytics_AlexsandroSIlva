# Dataset

```python
#caminho do dataset
caminho_arquivo = r"Dataset de Produtos de Beleza e Cuidados Pessoais.csv"
#leitura do dataset ele está separado por ; e seu encodign é o UTF
df = pd.read_csv(caminho_arquivo, sep=";", encoding='UTF-8')

#Para saber o numero de linhas e colunas que o dataset tem  
linhas, colunas = df.shape
#faz uma soma das linhas vazia que o dataset tem 
valores_vazios = df.isnull().sum().sum()
#imprimi as 20 primeiras linhas do dataset 
display(df.head(20))

#imprimi quantas linhas e colunas o dataset tem
print(f"\nO dataset possui {linhas} linhas e {colunas} colunas.")
#imprimi quantas linhas vazias o dataset tem
print(f"\nExistem {valores_vazios} valores vazios no dataset.")
```

```python
# Filtrando as linhas onde na coluna 'Preço' está vazia no caso o preço do produto
linhas_sem_preço_produto = df[df['Preço'].isnull()]

# imprimi as linhas sem o Preço do produto
print("Linhas sem o preço do Produto:")
#imprimi as 20 linhas do dataset que estão sem seu respectivo preço
display(linhas_sem_preço_produto.head(20))
```

# Limpeza e Organização de Dados

```python
#Primeiro vamos começar corrigindo as datas primeiro 
df['Data'] = df['Data'].str.replace('/2024', '/2023', regex=False)
#usando o replace para garanti que todas as datas sejam do ano certo 2023 no caso

#Agora vamos corrigir os Produtos que estão sem seus devidos ID
# Criando um dicionário de produtos e seus respectivos IDs
ids_correcoes = {
    "Batom": 34151,
    "Creme para Mãos": 35367,
    "Kit Matizador Shampoo e Condicionador": 77567,
    "Shampoo Matizador":77564

}

# Preenchendo esses IDS vazios
df['ID'] = df.apply(lambda row: ids_correcoes[row['Produto']] if pd.isnull(row['ID']) and row['Produto'] in ids_correcoes else row['ID'], axis=1)

#corrigindo os Produtos que estão sem seus devidos nome
produtos_correcoes = {
    16532: "Óleo Capilar"
}

# Preenchendo esse nome vazio
df['Produto'] = df.apply(lambda row: produtos_correcoes[row['ID']] if pd.isnull(row['Produto']) and row['ID'] in produtos_correcoes else row['Produto'], axis=1)

#como os valores do preço foram salvos no excel desse jeito 20,99 com virgula (padrão Brazileiro) e o Python entende esses valores (padrão americano) com . desse jeito 20.99 temos que converter agora para o Python consiga reconhecer e preeencher corretamente
produtos_faltando_preco = df[df['Preço'].isna()]['Produto'].unique()
df['Produto'] = df['Produto'].astype(str).str.strip()

#corrigindo os Produtos que estão sem seus devidos Preços
precos_correcoes = {
    "Shampoo": 18.50,
    "Batom Líquido": 27.99
}

# Preenchendo esses Preços vazios
df['Preço'] = df.apply(lambda row: precos_correcoes[row['Produto']] if pd.isnull(row['Preço']) and row['Produto'] in precos_correcoes else row['Preço'],axis=1)

#Convertendo a coluna inteira dos preços do formato brasileiro (com vírgulas) para o americano (com pontos)
df['Preço'] = df['Preço'].astype(str).str.replace(',', '.').astype(float)

# Removendo as duplicatas com base nas colunas 'Data' e 'Produto'
df_limpo = df.drop_duplicates(subset=['Data', 'Produto'])


#Para saber o numero de linhas e colunas que o dataset limpo tem 
linhas, colunas = df_limpo.shape

#imprimi as 30 primeiras linhas do dataset limpo
display(df_limpo.head(30))

#imprimi quantas linhas e colunas o dataset limpo tem agora
print(f"\nO dataset limpo possui {linhas} linhas e {colunas} colunas.")

# Salvando o dataset limpo 
df_limpo.to_csv('data_clean.csv', index=False, encoding='latin1')

print("Dataset limpo salvo como 'data_clean.csv'")


```

# Análise de Dados de Vendas

```python
#caminho do dataset_limpo
caminho_arquivo_limpo = r"data_clean.csv"
#leitura do dataset ele está separado por ; e seu encodign é o UTF
df_limpo = pd.read_csv(caminho_arquivo_limpo, sep=",", encoding='latin1')
```

```python
#tive que dropar uma linha que foi reconhecida como naN, é depois da ultima linha do dataset então é normal ela ser zerada
df_limpo = df_limpo.dropna(subset=['Quantidade', 'Preço'])

# Calculando o total de vendas (Quantidade * Preço) por produto
df_limpo['Total_Vendas'] = df_limpo['Quantidade'] * df_limpo['Preço']

# Agrupando por produto para calcular o total de vendas por produto
vendas_por_produto = df_limpo.groupby('Produto')['Total_Vendas'].sum().reset_index()

# O produto com o maior número de vendas totais
produto_mais_vendido = vendas_por_produto.loc[vendas_por_produto['Total_Vendas'].idxmax()]

# Imprimindo os resultados do total de vendas de todos os produtos
print("Total de vendas por produto:")
display(vendas_por_produto)

# Imprimindo os resultados do total de vendas de todos os produtos
print("\nProduto com o maior número de vendas totais:")
print(produto_mais_vendido)
```
